```python
import json

def format_sample(sample: dict) -> dict:
    patient = sample["input"]["patient"]
    diagnosis = sample["input"]["diagnosis"]
    context = sample["input"]["clinical_context"]
    
    # Собираем input в читаемую строку
    input_text = f"""Пациент: {patient['age']} лет, пол: {patient['sex']}, курение: {patient['smoking_history']}, ECOG: {patient['ECOG']}.
Диагноз: {diagnosis['type']}, {diagnosis['histology']}, стадия {diagnosis['stage']}, TNM: {diagnosis['TNM']}.
Метастазы: {', '.join(diagnosis['metastases'])}.
Молекулярный профиль: EGFR {diagnosis['EGFR']}, ALK {diagnosis['ALK']}, ROS1 {diagnosis['ROS1']}, BRAF {diagnosis['BRAF']}, PD-L1 {diagnosis['PD-L1']}.
Симптомы: {', '.join(context['symptoms'])}.
Предыдущее лечение: {context['previous_treatment']}."""

    return {
        "instruction": sample["instruction"],
        "input": input_text,
        "output": sample["output"]
    }
```

```python
def format_sample(sample: dict) -> dict:
    p = sample["input"]["patient"]
    d = sample["input"]["diagnosis"]
    c = sample["input"]["clinical_context"]

    input_text = (
        f"Пациент: {p['age']} лет, пол: {p['sex']}, "
        f"курение: {p['smoking_history']}, ECOG: {p['ECOG']}.\n"
        f"Диагноз: {d['type']}, {d['histology']}, стадия {d['stage']}, TNM: {d['TNM']}.\n"
        f"Метастазы: {', '.join(d['metastases'])}.\n"
        f"Молекулярный профиль: EGFR {d['EGFR']}, ALK {d['ALK']}, "
        f"ROS1 {d['ROS1']}, BRAF {d['BRAF']}, PD-L1 {d['PD-L1']}.\n"
        f"Симптомы: {', '.join(c['symptoms'])}.\n"
        f"Предыдущее лечение: {c['previous_treatment']}."
    )

    return {
        "text": f"<|im_start|>user\n{input_text}<|im_end|>\n"
                f"<|im_start|>assistant\n{sample['output']}<|im_end|>"
    }

dataset = load_dataset("json", data_files="../data/synthetic_data.json", split="train")
dataset = dataset.map(format_sample, remove_columns=dataset.column_names)
```

```python
dataset = dataset.train_test_split(test_size=0.1, seed=42)

train_dataset = dataset["train"]
test_dataset = dataset["test"]
```

```python
class LossVisualizationCallback(TrainerCallback):
    def __init__(self):
        self.steps = []
        self.losses = []
    
    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs and 'loss' in logs:
            self.steps.append(state.global_step)
            self.losses.append(logs['loss'])
            print(f"Step {state.global_step}: Loss = {logs['loss']:.4f}")
    
    def plot(self):
        import matplotlib.pyplot as plt
        plt.figure(figsize=(10, 6))
        plt.plot(self.steps, self.losses, 'b-', linewidth=2)
        plt.xlabel('Step')
        plt.ylabel('Training Loss')
        plt.title(f'Training Loss Over Steps (Final: {self.losses[-1]:.4f})')
        plt.grid(True, alpha=0.3)
        plt.show()
```

```python
def run_training(train_dataset, model_name="Qwen/Qwen2.5-0.5B-Instruct"):
    
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16,
        device_map="auto",
        trust_remote_code=True
    )
    
    tokenizer = AutoTokenizer.from_pretrained(
        model_name,
        trust_remote_code=True
    )
    tokenizer.pad_token = tokenizer.eos_token
    
    lora_config = LoraConfig(
        r=16,
        lora_alpha=32,
        target_modules=["q_proj", "v_proj", "k_proj", "o_proj"],
        lora_dropout=0.05,
        bias="none",
        task_type="CAUSAL_LM"
    )
    
    model = get_peft_model(model, lora_config)
    model.print_trainable_parameters()
    
    loss_callback = LossVisualizationCallback()
    
    with tempfile.TemporaryDirectory() as tmpdir:
        training_args = SFTConfig(
            output_dir=tmpdir,
            num_train_epochs=10,
            per_device_train_batch_size=1,
            gradient_accumulation_steps=4,
            learning_rate=2e-4,
            fp16=True,
            logging_steps=1,
            save_steps=50,
            max_seq_length=2048,
            dataset_text_field="text",
            report_to="none",
            save_strategy="no",
        )
        
        trainer = SFTTrainer(
            model=model,
            train_dataset=train_dataset,
            args=training_args,
            tokenizer=tokenizer,
        )
        
        trainer.add_callback(loss_callback)
        trainer.train()
        loss_callback.plot()
    
    return model, tokenizer


def cleanup(model=None, tokenizer=None):
    if model is not None:
        del model
    if tokenizer is not None:
        del tokenizer
    gc.collect()
    torch.cuda.empty_cache()
    print(f"GPU память: {torch.cuda.memory_allocated() / 1024**2:.1f} MB занято")
```

```python
model, tokenizer = run_training(train_dataset)
```

```python
cleanup(model, tokenizer)
```

```python
def test_model(model, tokenizer, sample_text):
    # Берём только часть до assistant — это и есть промпт
    prompt = sample_text.split("<|im_start|>assistant")[0] + "<|im_start|>assistant\n"
    
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=512,
            temperature=0.7,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )
    
    response = tokenizer.decode(
        outputs[0][inputs["input_ids"].shape[1]:], 
        skip_special_tokens=True
    )
    return response
```

```python
print(test_model(model, tokenizer, test_dataset[0]["text"]))
```

```python
base_model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen2.5-0.5B-Instruct",
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)
base_tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-0.5B-Instruct")
base_tokenizer.pad_token = base_tokenizer.eos_token

test_sample = test_dataset[0]["text"]

print(test_model(base_model, base_tokenizer, test_sample))
```
